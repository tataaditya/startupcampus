Mengobati ayam yang terkena Salmonella memerlukan pendekatan yang hati-hati, dan biasanya, pengobatan terbaik akan berkonsultasi dengan dokter hewan. Namun, saya dapat memberikan beberapa informasi umum mengenai jenis obat yang biasa digunakan untuk mengatasi infeksi bakteri seperti Salmonella pada ayam.

1. **Antibiotik**: Antibiotik sering digunakan untuk mengobati infeksi Salmonella. Beberapa jenis antibiotik yang umum digunakan termasuk:
   - **Enrofloxacin**: Dijual dengan merk seperti Baytril. Harga rata-rata untuk Baytril bervariasi tergantung pada wilayah dan dosis, tetapi bisa berkisar antara Rp 100.000 hingga Rp 500.000 per botol kecil (biasanya 100 ml).
   - **Amoxicillin**: Dikenal dengan berbagai merk, seperti Amoxivet. Harga bisa berkisar antara Rp 50.000 hingga Rp 150.000 per botol.
   - **Tetracycline**: Dijual dengan merk seperti Oxyvet. Harganya bervariasi, sekitar Rp 50.000 hingga Rp 100.000 per paket.

2. **Probiotik dan Suplemen**: Setelah pengobatan antibiotik, pemberian probiotik dapat membantu memulihkan keseimbangan mikroflora usus ayam.
   - **Protexin**: Sebuah probiotik yang dapat membantu pemulihan setelah antibiotik. Harga rata-rata sekitar Rp 100.000 hingga Rp 200.000 per paket.

3. **Vaksinasi**: Mencegah lebih baik daripada mengobati. Vaksinasi terhadap Salmonella dapat dilakukan untuk mencegah infeksi di masa depan. Harga vaksin bervariasi tergantung pada jenis dan pabrikannya.

Penting untuk diingat bahwa penggunaan antibiotik harus dilakukan dengan bijaksana dan sesuai petunjuk dokter hewan karena penggunaan yang tidak tepat dapat menyebabkan resistensi antibiotik. Selain itu, pastikan untuk mengikuti petunjuk dosis dan periode penarikan sebelum ayam dapat dikonsumsi.

Sebaiknya konsultasikan terlebih dahulu dengan dokter hewan untuk mendapatkan diagnosis yang tepat dan rekomendasi pengobatan yang sesuai.

```python
from langchain_openai import OpenAIEmbeddings
import pandas as pd
import numpy as np
from openai import OpenAI

df = pd.read_csv('example_input.csv', delimiter = ";")
df.head()
```

```python
client = OpenAI()
def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
df['embedding'] = df.text_input.apply(lambda x: get_embedding(x, model='text-embedding-3-small'))
df.head()
```
